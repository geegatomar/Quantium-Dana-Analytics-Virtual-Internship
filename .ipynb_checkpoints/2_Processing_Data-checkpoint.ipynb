```python
adf = df[['DATE', 'STORE_NBR', 'PROD_QTY', 'TOT_SALES']]
```

```python
adf
```

```python
adf.groupby(['STORE_NBR', df['DATE'].dt.strftime('%B')]).first()
```

```python
# We could have written df['DATE'].dt.month to do the same,
# But we did df['DATE'].dt.strftime('%B %Y') where %B is for month, %Y is for year
# This gives more written detailed grouping on representation
```

```python
aggreg = {
    #'DATE': 'first',
    'PROD_QTY': sum,
    'TOT_SALES': sum
}

grouped = adf.groupby(['STORE_NBR', df['DATE'].dt.strftime('%B %Y')]).agg(aggreg)
```

```python
from scipy.stats import pearsonr

l1 = grouped.loc[(1, ), :].PROD_QTY
l77 = grouped.loc[(77, ), :].PROD_QTY

pearsonr(l1, l77)
```

```python
store77 = []
store86 = []
store88 = []

s = grouped.loc[(1, ), :].TOT_SALES
s77 = grouped.loc[(77, ), :].TOT_SALES
s86 = grouped.loc[(86, ), :].TOT_SALES
s88 = grouped.loc[(88, ), :].TOT_SALES
    
    
l = grouped.loc[(1, ), :].PROD_QTY
l77 = grouped.loc[(77, ), :].PROD_QTY
l86 = grouped.loc[(86, ), :].PROD_QTY
l88 = grouped.loc[(88, ), :].PROD_QTY

print(pearsonr(s, s77)[0], pearsonr(l, l77)[0])

store77.append((pearsonr(s, s77)[0], pearsonr(l, l77)[0]))
store86.append((pearsonr(s, s86)[0], pearsonr(l, l86)[0]))
store88.append((pearsonr(s, s88)[0], pearsonr(l, l88)[0]))

print(store77)
print(store86)
print(store88)
```

```python
import numpy as np

store77 = []
store86 = []
store88 = []

for store in range(1, 273) :#and store not in {77, 86, 88}:
    if store in {77, 86, 88}:
        continue
    s = grouped.loc[(store, ), :].TOT_SALES
    s77 = grouped.loc[(77, ), :].TOT_SALES
    s86 = grouped.loc[(86, ), :].TOT_SALES
    s88 = grouped.loc[(88, ), :].TOT_SALES
    
    
    l = grouped.loc[(store, ), :].PROD_QTY
    l77 = grouped.loc[(77, ), :].PROD_QTY
    l86 = grouped.loc[(86, ), :].PROD_QTY
    l88 = grouped.loc[(88, ), :].PROD_QTY
    
    try:
        store77.append((store, pearsonr(s, s77)[0], pearsonr(l, l77)[0]))
        store86.append((store, pearsonr(s, s86)[0], pearsonr(l, l86)[0]))
        store88.append((store, pearsonr(s, s88)[0], pearsonr(l, l88)[0]))
    except:
        pass
    
```
